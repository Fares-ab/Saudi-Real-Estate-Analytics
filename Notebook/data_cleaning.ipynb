```python
import pandas as pd

file_path = "../Data/cleaned_real_estate_listings.csv"

df = pd.read_csv(
    file_path,
    encoding="utf-8-sig"
)

print("Rows, Columns:")
print(df.shape)

print("\nColumns:")
print(df.columns.tolist())

df.head()
```

```python
df.info()
```

```python
missing_values = df.isnull().sum().sort_values(ascending=False)
missing_values
```

```python
df[
    [
        "price",
        "area",
        "beds",
        "livings",
        "bathrooms",
        "street_width",
        "age"
    ]
].describe()
```

```python
df_clean = df.copy()

df_clean.head()
```

```python
print(df_clean.shape)
df_clean.info()
```

```python
df_clean[["price", "area"]].describe()
```

```python
df_clean = df_clean[
    (df_clean["price"] > 1000) &
    (df_clean["price"] < 100000000) &
    (df_clean["area"] > 20) &
    (df_clean["area"] < 100000)
].copy()

print(df_clean.shape)

df_clean[["price", "area"]].describe()
```

```python
df_clean["price_per_sqm"] = df_clean["price"] / df_clean["area"]

df_clean[["price", "area", "price_per_sqm"]].describe()
```

```python
pd.set_option("display.float_format", "{:,.2f}".format)
```

```python
df_clean = df_clean[
    (df_clean["price_per_sqm"] >= 100) &
    (df_clean["price_per_sqm"] <= 50000)
].copy()

print(df_clean.shape)

df_clean[["price", "area", "price_per_sqm"]].describe()
```

```python
def classify_listing_type(title):
    title = str(title)
    
    if "للبيع" in title:
        return "Sale"
    elif "للإيجار" in title or "للايجار" in title or "ايجار" in title:
        return "Rent"
    else:
        return "Unknown"

df_clean["listing_type"] = df_clean["title"].apply(classify_listing_type)

df_clean["listing_type"].value_counts()
```

```python
df_clean = df_clean[df_clean["listing_type"] != "Unknown"].copy()

df_clean["listing_type"].value_counts()
```

```python
df_clean["create_time"].head()
```

```python
df_clean["create_date"] = pd.to_datetime(df_clean["create_time"], unit="s")

df_clean[["create_time", "create_date"]].head()
```

```python
df_clean["year"] = df_clean["create_date"].dt.year
df_clean["month"] = df_clean["create_date"].dt.month
df_clean["year_month"] = df_clean["create_date"].dt.to_period("M").astype(str)

df_clean[["create_date", "year", "month", "year_month"]].head()
```

```python
df_clean["create_date"].min(), df_clean["create_date"].max()
```

```python
df_clean.isnull().sum().sort_values(ascending=False)
```

```python
# Drop columns with too many missing values or low analytical value
columns_to_drop = ["rent_period", "type", "length", "width"]

df_clean = df_clean.drop(columns=columns_to_drop)

# Fill missing values for property room-related columns
df_clean["beds"] = df_clean["beds"].fillna(0)
df_clean["wc"] = df_clean["wc"].fillna(0)
df_clean["livings"] = df_clean["livings"].fillna(0)

# Fill missing advertiser type as Unknown
df_clean["advertiser_type"] = df_clean["advertiser_type"].fillna("Unknown")

```

```python
df_clean.isnull().sum().sort_values(ascending=False)
```

```python
df_clean.duplicated().sum()
```

```python
df_clean["id"].duplicated().sum()
```

```python
df_clean = df_clean.rename(columns={
    "id": "listing_id",
    "location.lat": "latitude",
    "location.lng": "longitude",
    "wc": "bathrooms"
})

df_clean.columns.tolist()
```

```python
final_columns = [
    "listing_id",
    "title",
    "listing_type",
    "category",
    "city",
    "district",
    "price",
    "area",
    "price_per_sqm",
    "beds",
    "livings",
    "bathrooms",
    "street_width",
    "age",
    "latitude",
    "longitude",
    "advertiser_type",
    "create_time",
    "create_date",
    "year",
    "month",
    "year_month"
]

df_clean = df_clean[final_columns]

df_clean.head()
```

```python
print(df_clean.shape)
df_clean.info()
```

```python
df_clean = df_clean.rename(columns={
    "year": "listing_year",
    "month": "listing_month",
    "year_month": "period_month"
})

print("Columns renamed successfully")
print(df_clean.columns.tolist())
```

```python
output_path = "../Data/cleaned_real_estate_listings.csv"

df_clean.to_csv(
    output_path,
    index=False,
    encoding="utf-8-sig"
)

print("Cleaned file saved successfully")
```
